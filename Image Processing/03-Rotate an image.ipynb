## Rotate an image

```python
from PIL import Image  
import matplotlib.pyplot as plt
img=Image.open('./image/tyrion_lannister.jpg')

plt.imshow(img)
plt.show()
```

```python
img_rotate = img.rotate(45)  


plt.imshow(img_rotate)
plt.show()
```

```python
img_rotate = img.transpose(Image.ROTATE_90) 

plt.imshow(img_rotate)
plt.show()
```
