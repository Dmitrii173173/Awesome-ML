## Reading and Inverting an Image Using Python

The necessary python libraries

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import PIL.Image as Image 
```

The following function computes the negative of a given image. It accepts the original image (in the form of a matrix) as input and returns its negative as output.

```python
def computeNegativeImage(imge):
    """
    Computes the negative pixel values of an image.
    
    Parameters
    ----------
    imge : ndarray
        Input array that stores the image pixel values.
    
    Returns
    -------
    negImge : ndarray
        The negative pixel values of the original image.
    """

    # Compute the absolute of the resulting values
    negImge = np.abs(imge - 255)
    
    # Return the final result
    return negImge      
```

## Testing

```python
#Read an image file
imge = Image.open('./image/skynews-cersei.jpg') # open an image

#Convert the image file to a matrix
imge = np.array(imge)

#Convert the uint datatype of the matrix values into 'int' for using the negative values
imge = imge.astype('int')

##Note: If the image is RGB colored, we can convert to grayscale image as follows 
#imge = imge.convert(mode='L')
```

To check the dimension(size) of the image:

```python
print ("Image Size: ",imge.shape)
```

## The original image before it is converted.

```python
#plt.figure(figsize=(5,5))
plt.imshow(imge, cmap=plt.get_cmap('gray'))
plt.title('Original Image')
plt.show()
```

The negative of the image can be computed by using the previously constructed function. We can also show it using the same way as above.

```python
#Compute the negative of the given image.
negImge = computeNegativeImage(imge)

#Display the resulting image
plt.imshow(negImge, cmap=plt.get_cmap('gray'))
plt.title('Negative of the Image')
plt.show()
```

The given result shows that all of the colors are interchanged. We can see the impact of the process by comparing the black and white colors in the original and final photographs.
