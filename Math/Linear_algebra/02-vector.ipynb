## Vector

```python
import numpy as np

# Define two vectors
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Calculate the projection of vector a onto vector b
projection = np.dot(a, b) / np.dot(b, b) * b

print("Vector a:", a)
print("Vector b:", b)
print("Projection of a onto b:", projection)
```
